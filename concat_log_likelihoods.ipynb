# Concat pointwise log likelihoods

```python
import numpy as np
import xarray as xr
import arviz as az
rng = np.random.default_rng()
```

```python
letters = "abcde"
shapes = (15, 23, 31, 19, 27)
dims = {letter: [f"{letter} row"] for letter in letters}
data = {letter: rng.normal(size=(4, 500, shape)) for letter, shape in zip(letters, shapes)}
ds = az.dict_to_dataset(data, dims=dims)
ds
```

```python
ds["log_likelihood"] = xr.concat(
    (ds[letter].rename({f"{letter} row": "row"}) for letter in letters), 
    dim="row"
)
ds
```
